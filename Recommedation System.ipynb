```python
def important_features(dataset):
    data=dataset.copy()
    for i in range(0, dataset.shape[0]):
        data["imp"]=data["title"]+' '+data["description"]+" "+data["categories"]
    return data
```

```python
data=important_features(dataset)
```

```python
def recommend(title):
    product_id=data[data.title==title]["ids"].values[0]
    print("product_id", product_id)
    scores=list(enumerate(sim[product_id]))
  #  print("scores", scores)
    sorted_scores=sorted(scores,key=lambda x:x[1],reverse=True)
   # print("sorted_scores", sorted_scores)
    sorted_scores=sorted_scores[1:]
    # categoryproducts = 
    
    # categoryproducts=[data[products[0]==data["ids"]]["categories"].values[0] for products in sorted_scores]
    # categoryproduct = [data[tempt[0]==data["categories"][product_id]]["title"].values[0] for tempt in categoryproducts]
    sorted_Score_By_Categories = []
    for products in sorted_scores:
      catogoryEachProduct = data[products[0]==data["ids"]]["categories"].values[0]   
      if catogoryEachProduct == data["categories"][product_id]:
        sorted_Score_By_Categories.append (products)
    sorted_Title=[data[products[0]==data["ids"]]["title"].values[0] for products in sorted_Score_By_Categories]      
    return sorted_Title
```

```python
def recommend_ten(product_list):
    first_ten=[]
    count=0
    for product in product_list:
        if count > 9:
            break
        count+=1
        first_ten.append(product)
    return first_ten
```

```python
lst=recommend("Energizer 2 in 1 Light")
#print ("lst", lst)
m=recommend_ten(lst)
```
